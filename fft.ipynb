```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq
from scipy.io import wavfile as wav
import os

# Plot and save the Fourier transform of audio file `filename` to `images/`
def compute_fft(filename):
    rate, data = wav.read('sounds/'+filename)
    data = data[:,0]
    samples = data.shape[0]
    datafft = fft(data)
    fftabs = abs(datafft)
    freqs = fftfreq(samples,1/rate)

    plt.figure()
    plt.xlim( [10, rate/2] )
    plt.xscale( 'log' )
    plt.grid( True )
    plt.xlabel( 'Frequency (Hz)' )
    plt.title(filename)
    plt.plot(freqs[:int(freqs.size/2)],fftabs[:int(freqs.size/2)])

    name = os.path.splitext(filename)[0]
    plt.savefig('images/'+name+'.png')

    
for filename in os.listdir('sounds'):
    ext = os.path.splitext(filename)[1]
    if ext in ['.wav']:
        compute_fft(filename)
```

```python
compute_fft('xylophone-a.wav')
```
