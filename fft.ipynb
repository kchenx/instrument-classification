```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq
from scipy.io import wavfile as wav

def plot_fft(filename):
    rate, data = wav.read('audiofiles/'+filename)
    data = data[:,0]
    samples = data.shape[0]
    datafft = fft(data)
    fftabs = abs(datafft)
    freqs = fftfreq(samples,1/rate)

    plt.xlim( [10, rate/2] )
    plt.xscale( 'log' )
    plt.grid( True )
    plt.xlabel( 'Frequency (Hz)' )
    plt.plot(freqs[:int(freqs.size/2)],fftabs[:int(freqs.size/2)])
```

```python
plot_fft('piano-a.wav')
```

```python
plot_fft('xylophone-a.wav')
```
